```python
import cv2
import os
import glob
import os.path
```

```python
dir = "./dataset/videos"
files = glob.glob(os.path.join(dir, '*.mp4'))
print(files)
```

```python
try: 
    # creating a folder named data
    if not os.path.exists('./dataset/all'):
        os.makedirs("./dataset/all")

# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')
        
```

```python
# image-counter
image_counter = 0

# Read the videos from specified path
for video_file in files:
    video_data = cv2.VideoCapture(video_file)
  
    while(True):
        # reading from frame
        ret,frame = video_data.read()

        if ret:
            # if video is still left continue creating images
            name = './dataset/all/image_' + str(image_counter) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            image_counter += 1
        else:
            break

    # Release all space and windows once done
    video_data.release()
cv2.destroyAllWindows()
```
